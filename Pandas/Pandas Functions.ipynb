# Handling Missing Value

```python
df = pd.read_csv('data2.csv')
df.head()
```

```python
df.isnull()   #find null values
```

```python
df.isnull().sum()   #summary of null values
```

```python
df.notnull()   
```

```python
df.notnull().sum()
```

# dropna()

```python
df.dropna()  #drop the null value rows or df.dropna(how ='any')
```

```python
df.dropna(axis = 1)  #drop the null values cols or df.dropna(axis=1,how='any')
```

```python
df = pd.read_csv('data2.csv')
df.head()
```

```python
df.dropna(thresh=8)  #rows must contain 8 not null values
```

```python
df.dropna(inplace=True) #inplace will update the same dataframe
df
```

# fillna()

```python
df = pd.read_csv('data2.csv')
df.head()
```

```python
df.fillna(0)   #nan value will fill with 0
```

```python
df.fillna({'Industry': 'none','Expenses' : 0 })  #specific value for specific cols
```

```python
df.fillna(method ='ffill')  #null value will fill with previous rows value
```

```python
df.fillna(method ='bfill')  #null value will fill with next rows value
```

# replace()

```python
df = pd.read_csv('data2.csv')
df.head()
```

```python
df.replace( 'Health', 'Finance')  # will replace 'Health' value with 'Finance'
```

```python
df.replace( 'Health', method = 'ffill')  #replace with previous rows values
```

```python
#Merging Function
df1 = pd.DataFrame({'ID': [1,2,3,4]
                   ,'CLASS':[9,10,11,23]})
df1
```

```python
df2 = pd.DataFrame({'ID': [1,2,3,4]
                   ,'NAME':['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2,on ='ID')
```

```python
#concat function
pd.concat([df1,df2])
```

```python
#join function
df1.join(df2, lsuffix ='_1')  #ID is similar so to join it have to change one
```
